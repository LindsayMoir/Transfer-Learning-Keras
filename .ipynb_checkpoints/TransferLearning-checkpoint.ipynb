```python
# dataset paths
train_path = r'C:\Users\Lindsay\AI_P6\photos\train'
test_path = r'C:\Users\Lindsay\AI_P6\photos\test'
```

```python
# Get the path name of photos
files = glob.glob(test_path + r'\drink\*')
len(files)
```

```python
# Instantiate vgg
vgg = VGG16(input_shape = image_size + [3], 
            weights = 'imagenet', 
            include_top = False)
```

```python
# Look at the input tensor
vgg.input
```

```python
# Tell vgg that it's layers ARE NOT adjustable.
for layer in vgg.layers:
  layer.trainable = False
```

```python
# Get the number of folders which is the same as the number of labels
folders = glob.glob(r'C:\Users\Lindsay\AI_P6\photos\train\*')
print(len(folders))
```

```python
# Flatten the output
x = Flatten()(vgg.output)

# Put the final output layer on
prediction = Dense(len(folders), 
                   activation = 'softmax')(x)

# Build final model
model = Model(inputs = vgg.input, 
              outputs = prediction)

model.summary()
```

```python
# set hyperparameters
adam = optimizers.Adam()
model.compile(loss = 'binary_crossentropy',
              optimizer = adam,
              metrics = ['accuracy'])
```

```python
# Create the train generator
train_datagen = ImageDataGenerator(
    preprocessing_function = preprocess_input,
    rotation_range = 40,
    width_shift_range = 0.2,
    height_shift_range = 0.2,
    shear_range = 0.2,
    zoom_range = 0.2,
    horizontal_flip = True,
    fill_mode = 'nearest')
```

```python
# create the test generator
test_datagen = ImageDataGenerator(
    preprocessing_function = preprocess_input,
    rotation_range = 40,
    width_shift_range = 0.2,
    height_shift_range = 0.2,
    shear_range = 0.2,
    zoom_range = 0.2,
    horizontal_flip = True,
    fill_mode = 'nearest')
```

```python
train_set = train_datagen.flow_from_directory(train_path,
                                              target_size = (100, 100),
                                              batch_size = 32,
                                              class_mode = 'categorical')
```

```python
test_set = test_datagen.flow_from_directory(test_path,
                                            target_size = (100, 100),
                                            batch_size = 32,
                                            class_mode = 'categorical')
```

```python
# Setup the model and fit it

checkpoint = ModelCheckpoint(filepath = 'mymodel.h5', 
                             verbose = 2, 
                             save_best_only = True)

callbacks = [checkpoint]

start = datetime.now()

model_history = model.fit(
    train_set,
    validation_data = test_set,
    epochs = 6,
    steps_per_epoch = 1024,
    validation_steps = 96,
    callbacks = callbacks,
    verbose = 1)

duration = datetime.now() - start
print("Training completed in time: ", duration)
```

```python
# Plot training & validation loss values
plt.plot(model_history.history['accuracy'])
plt.plot(model_history.history['val_accuracy'])
plt.title('CNN Model accuracy values')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend(['Train', 'Test'], loc = 'upper left')
plt.show()
```
